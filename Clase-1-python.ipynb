# Estructuras de control
Veamos las estructuras de control son palabras reservadas con las cuales le vamos a estar indicando a python que realice una serie de operaciones. La primera es **IF**

```python
if True:
    print('Es verdadero')

if False:
    print('Es falso')
```

Vemos que solo imprime en pantalla lo que está dentro del primer if (asi es como "metemos" cosas dentro de una estructura de control, simplemente con TAB hacia la derecha). Aqui lo que le dijimos es que evalúe si la condición era verdadera en cada if, logicamente solo la primera fue verdadera por ponerle literalmente "verdad" en ese lugar. Ahora podemos hacer cosas mas interesantes como comparaciones

```python
if 1 == 1:
    print('Es verdadero 1 == 1')

if 1 == 2:
    print('Es verdadero 1 == 2 ')
```

Tambien operaciones con mayor, menor, divisiones con resto y divisiones solo su parte entera:

```python
if 1 < 2 :
    print('1 < 2 es verdadero')

if 4 < 2 :
    print('4 < 2 es verdadero')

if 1//2 == 0:
    print('1//2 == 0 es verdadero')

if 4//3 == 1:
    print('4//3 == 1 es verdadero')

if 4%3 == 1:
    print('4%3 == 1 es verdadero')

if 'a' in 'hola':
    print('a está dentro de hola, verdadero')

if 4 in {1,2,3}:
    print('1 está dentro del set')
else:
    print('4 no está dentro del set')
```

## Bucles
Los bucles son estructuras de control extremadamente útiles, ya que es el corazón de los lenguajes de control se podria decir. Gracias a ellos es que la programación es lo que es. Su lógica trata de iterar n cantidad de veces según condiciones lógicas.

### FOR 
Para el for al menos en python, lo que se hace es darle una variable **iterable** y for la recorrerá de principio a fin. Esto es un poco distinto a como lo tomamos de otros lenguajes de programación, ya que en otros lenguajes se utiliza un condicional, sin embargo aquí en python es de la siguiente manera

```python
lista = [1,2,3,4,5,6]

for i in lista:
    print(i)
```

En este caso por ejemplo pasamos por todos los elementos de la lista.

```python
for i in lista[1::2]:
    print(i)
```

Para este otro caso pasamos por todos los elementos de la lista pero de la siguiente manera:
[A:B:C]
En donde A es donde empezamos, B es donde terminamos y C es cada cuantos leemos.
En este ejemplo anterior empezamos en 1, no decimos en donde terminamos (es decir hasta el final) y vamos leyendo cada dos números.

```python
for i in lista[1:6:2]:
    print(i)
```

Aqui nuevamente hacemos lo mismo pero declarando **hasta** donde vamos a leer:
comenzamos en el 1, leemos hasta el 6 y leemos cada 2 números.

```python
for i in lista[6:0:-2]:
    print(i)
```

Aquí otro ejemplo interesante, esta vez comenzamos en el 6, vamos hasta el 0 y leemos cada -2. Esto quiere decir que vamos a leer cada dos números pero yendo para atras, es decir disminuyendo 2 por cada lectura.

## While
El while si es con un condicional mas clásico en comparación con los otros lenguajes de programación y en este caso lo que hacemos es repetir infinitamente **hasta** que lleguemos a que la condición del while sea completamente falsa.

```python
counter = 0
while counter < 5:
    counter += 1
    print(counter)
```

Este tipo de iteración es **muy** poderosa sin embargo también es peligrosa, ya que si nos quedamos en una condición que es **eternamente verdad** jamas podremos salir del while, lo cual en un programa, generalmente está mal.
