## Filtrado de datos con condicionales encadenados

Vamos a poner la condición de que la temperatura sea mayor a 20 grados y a su vez menor a 35 grados

```python
# Le estoy diciendo que la temperatura sea mayor a 20 y menor a 35
df_filtrado = df[(df['Temp [°C]'] > 20) & (df['Temp [°C]'] < 35)] 
df_filtrado
```

Ahora vamos a poner un filtro con la condición "O", esta vez vamos a pedir que la temperatura sea menor a 20 grados y mayor a 35 grados

```python
# Le estoy diciendo que la temperatura sea mayor a 20 o menor a 35
df_filtrado = df[(df['Temp [°C]'] < 20) | (df['Temp [°C]'] > 35)] 
df_filtrado
```

# Nuestro primer modelo de Machine Learning - Ajuste lineal

En este caso vamos a estar trabajando con los mismos datos que antes (datos reales) en los cuales vamos a realizar un ajuste lineal con el modelo de regresión lineal. Este modelo proviene de una biblioteca que no es escencialmente de machine learning, pero si es de modelos estadisticos lo cual (en escencia) es lo mismo, puesto que los modelos en el fondo son eso, modelos estadísticos.

```python
import numpy as np
import matplotlib.pyplot as plt # Biblioteca gráfica
from scipy.stats import linregress # Biblioteca que trae el modelo lineal de Machine Learning
import pandas as pd # Importamos pandas

# Lee el archivo de Excel
df = pd.read_excel("OPD_vs_Temp_serie_D6C.xlsx")

# Extrae los valores de las columnas en vectores
temperatura = df['OPD [nm]'].values # Solo dame los datos de la columna y NO SUS INDICES.
opd = df['Temp [°C]'].values # Solo dame los datos de la columna y NO SUS INDICES.

print(f'\n\n\nLa cantidad de mediciones hechas fueron: {len(df)}\n\n\n');

# Parámetros agregados de estílo.
# Use LaTeX-style font with serif
plt.rcParams.update({
    'font.family': 'serif',
    'text.usetex': True,  # Enables LaTeX rendering
    'font.size': 18,  # Adjust overall font size
    'axes.labelsize': 18,  # Font size for axis labels
    'xtick.labelsize': 18,  # Font size for x-axis ticks
    'ytick.labelsize': 18,  # Font size for y-axis ticks
})

# Crear una figura más grande
plt.figure(figsize=(12, 8))  # Aumenta tamaño del gráfico

# Definir errores
error_temperatura    = 0.6;
error_nanometros_opd = 3;

# Graficar puntos con barras de error
#plt.errorbar(temperatura, opd, xerr=error_temperatura, yerr=error_nanometros_opd, fmt='o', label='Datos experimentales')
plt.scatter(temperatura, opd)

# Realizar regresión lineal
# Esta linea es en la cual se realiza el ajuste lineal con los datos
slope, intercept, r_value, p_value, std_err = linregress(temperatura, opd) 
# Slope: PEndiente de la recta
# Intercept: es la ordenada de origen,
# r_value: Es el coeficiente de relación lineal, mientras mas alto, mas lineal es el sistema.
# p_value: es el valor asociado al prueba estadistica.
# std_err: Error estandar de la pendiente.

# Calcular los valores ajustados (recta)

opd_ajustada = intercept + slope * temperatura

# Graficar la recta de regresión original
plt.plot(temperatura, opd_ajustada, color='red', label=f'Linear regression: y = {slope:.4f}x + {intercept:.4f}')

# Actualizar la leyenda con m_med y m_error
plt.legend(title=f'm = {slope:.4f} ± {std_err:.4f} °C/nm')

# Etiquetas y título
plt.ylabel('Temperature (°C)')
plt.xlabel('OPD (nm)')
plt.title('Temperature vs OPD for series S2 of measures')
plt.grid(True)

# Mostrar gráfico
plt.savefig('figura.jpg')
plt.show()


# Imprimir los resultados
print(f'p_value: {p_value}')
print(f'std_err: {std_err}')
print(f'slope: {slope}')

```
